```python
# ===========================================
# IMPORTS
# ===========================================
from pathlib import Path
from datetime import datetime
import glob

import numpy as np
import pandas as pd

from sklearn.preprocessing import (
    StandardScaler,
    MinMaxScaler
)
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import silhouette_score

import matplotlib.pyplot as plt
import seaborn as sns

from fuzzywuzzy import process  # for possession-matching

# -------------------------------------------
# Matplotlib / seaborn styling
# -------------------------------------------
sns.set(style="whitegrid")

# ===========================================
# PATH SETUP
# Kodefilen ligger i prosjektroten:
# C:\Users\henri\Desktop\Fotballanalyser\Ball\projects\central-midfielders-clustering
# ===========================================
PROJECT_ROOT = Path(".").resolve()

DATA_RAW = PROJECT_ROOT / "data" / "raw"
DATA_EXAMPLE = PROJECT_ROOT / "data" / "example"
VIZ_DIR = PROJECT_ROOT / "viz"
VIZ_DIR.mkdir(parents=True, exist_ok=True)

# Toggle: bruk syntetisk eksempeldata eller ekte Wyscout-data
USE_EXAMPLE_DATA = False  # sett til True hvis du vil tvinge demo-modus

# ===========================================
# HJELPEFUNKSJON FOR Å LESE SESONGMAPPER
# ===========================================
def load_season_folder(folder: Path, pattern: str = "*.xlsx") -> pd.DataFrame:
    """Les alle Excel-filer i en sesongmappe (23/24 eller 25/26) og slå sammen til én DataFrame."""
    files = list(folder.glob(pattern))
    if not files:
        raise FileNotFoundError(
            f"Ingen filer funnet i {folder} med mønster {pattern}. "
            "Sjekk at du har lagt dine Wyscout-eksporter i riktig mappe."
        )
    dfs = [pd.read_excel(f) for f in files]
    return pd.concat(dfs, ignore_index=True)

# ===========================================
# LAST DATA (MED FALLBACK TIL EXAMPLE-FILE)
# ===========================================
wyscout_24_25_path = DATA_RAW / "wyscout_24_25.csv"
possession_path    = DATA_RAW / "df_possession.csv"
data_23_24_dir     = DATA_RAW / "23_24"
data_25_26_dir     = DATA_RAW / "25_26"
example_file       = DATA_EXAMPLE / "cm_fake_sample.csv"

df_24_25 = None
df_23_24 = None
df_25_26 = None
df_possession = None

if USE_EXAMPLE_DATA:
    # Tving demo-modus
    if not example_file.exists():
        raise FileNotFoundError(
            f"Fant ikke example-fil: {example_file}\n"
            "Legg inn cm_fake_sample.csv i data/example/ eller sett USE_EXAMPLE_DATA=False "
            "og legg inn wyscout_24_25.csv i data/raw/."
        )
    df_24_25 = pd.read_csv(example_file)
    print("Bruker syntetisk eksempeldata:", df_24_25.shape)
else:
    # Prøv å laste ekte 24/25-data
    if wyscout_24_25_path.exists():
        df_24_25 = pd.read_csv(wyscout_24_25_path)
        print("24/25:", df_24_25.shape)
    else:
        print(
            f"⚠️ Fant ikke {wyscout_24_25_path}. Faller tilbake til example-fil."
        )
        if not example_file.exists():
            raise FileNotFoundError(
                f"Verken {wyscout_24_25_path} eller example-fil {example_file} finnes.\n"
                "Legg inn minst én av dem for å kunne kjøre modellen."
            )
        df_24_25 = pd.read_csv(example_file)
        USE_EXAMPLE_DATA = True
        print("Bruker syntetisk eksempeldata:", df_24_25.shape)

    # Possession-data (valgfritt)
    if possession_path.exists():
        df_possession = pd.read_csv(possession_path, encoding="latin1")
        print("Possession-data:", df_possession.shape)
    else:
        print(f"⚠️ Possession-fil ikke funnet: {possession_path}. "
              "Modellen kjører uten possession-justering.")
        df_possession = None

    # 23/24 – flere filer (valgfritt, for robustness)
    if data_23_24_dir.exists():
        try:
            df_23_24 = load_season_folder(data_23_24_dir)
            print("23/24 total shape:", df_23_24.shape)
        except FileNotFoundError as e:
            print(f"⚠️ Robustness 23/24 hoppes over: {e}")
            df_23_24 = None
    else:
        print(f"⚠️ Mappe for 23/24 ikke funnet: {data_23_24_dir}")
        df_23_24 = None

    # 25/26 – flere filer (valgfritt, for robustness)
    if data_25_26_dir.exists():
        try:
            df_25_26 = load_season_folder(data_25_26_dir)
            print("25/26 total shape:", df_25_26.shape)
        except FileNotFoundError as e:
            print(f"⚠️ Robustness 25/26 hoppes over: {e}")
            df_25_26 = None
    else:
        print(f"⚠️ Mappe for 25/26 ikke funnet: {data_25_26_dir}")
        df_25_26 = None

# Sjekk at vi faktisk har 24/25-data (enten ekte eller example)
if df_24_25 is None:
    raise RuntimeError(
        "Ingen 24/25-data tilgjengelig (verken raw eller example). "
        "Modellen kan ikke trenes."
    )

# ===========================================
# PREPROCESSING-KLASSE FOR Wyscout-EXPORT
# ===========================================
class Preprocessing:
    def __init__(self, df, positions=None, min_minutes=None, leagues=None):
        """
        df: DataFrame som skal filtreres.
        positions: Liste med posisjoner (strenger) som skal være med (f.eks. ['RCMF','LCMF']).
        min_minutes: Minimum antall spilte minutter.
        leagues: Liste med liganavn som skal beholdes (strenger),
                 f.eks. ['Spain La Liga 2024-25', ...]. Hvis None -> ingen league-filter.
        """
        if positions is not None and not isinstance(positions, list):
            raise ValueError("positions must be a list or None")
        if leagues is not None and not isinstance(leagues, list):
            raise ValueError("leagues must be a list or None")

        self.df = df
        self.positions = positions
        self.min_minutes = min_minutes
        self.leagues = leagues
        self.filtered_df = None

    def filter_data(self):
        """Filtrerer DataFrame basert på angitte kriterier, men beholder alle Wyscout-kolonner."""
        df = self.df.copy()

        # 1) Gjør kolonnenavn lowercase
        df.columns = df.columns.str.lower()

        # 2) Normaliser lagnavn-kolonner
        rename_map = {}
        if 'team within selected timeframe' in df.columns:
            rename_map['team within selected timeframe'] = 'squad'
        if 'team' in df.columns:
            rename_map['team'] = 'club today'
        df = df.rename(columns=rename_map)

        # 3) League-filter (kun hvis både leagues er satt og 'league'-kolonnen finnes)
        if self.leagues is not None and 'league' in df.columns:
            df = df[df['league'].isin(self.leagues)]

        # 4) Konverter 'contract expires' → 'contract months left'
        if 'contract expires' in df.columns:
            df['contract expires'] = pd.to_datetime(df['contract expires'], errors='coerce')
            today = datetime.today()
            df['contract months left'] = df['contract expires'].apply(
                lambda x: 0 if pd.isna(x) else max(
                    0, int((x.year - today.year) * 12 + (x.month - today.month))
                )
            )
        else:
            df['contract months left'] = 0

        # 5) Filtrer posisjoner (hvis posisjoner er spesifisert)
        if self.positions is not None and 'position' in df.columns:
            pos_series = df['position'].fillna('')
            mask_positions = pos_series.apply(
                lambda x: any(pos in x for pos in self.positions)
            )
            df = df[mask_positions]

        # 6) Minutter-filter
        if self.min_minutes is not None and 'minutes played' in df.columns:
            df = df[df['minutes played'] > self.min_minutes]

        self.filtered_df = df
        return self.filtered_df

# ===========================================
# POSSESSION-JUSTERING
# ===========================================
class NormalizePossession:
    def __init__(self, df):
        """
        Initialiserer klassen med en kopi av DataFrame for å bevare originalen.
        Forutsetter at df har kolonnen 'poss_decimal' (lagbesittelse 0–1).
        """
        self.df = df.copy()

    def possession_adjust_def(self, value, team_possession):
        """Justerer en (volum)defensiv metric basert på motstanderens besittelse."""
        if pd.isnull(team_possession) or team_possession >= 1:
            return value
        opp_poss = 1 - team_possession
        return value if opp_poss == 0 else value * (0.50 / opp_poss)

    def possession_adjust_off(self, value, team_possession):
        """Justerer en (volum)offensiv metric basert på eget lags besittelse."""
        if pd.isnull(team_possession) or team_possession == 0:
            return value
        return value * (0.50 / team_possession)

    def apply_adjustments(self):
        """
        Justerer eksisterende kolonner direkte for ballbesittelse, uten å lage nye kolonner.
        """
        def_def_cols = [
            'defensive duels per 90', 'interceptions per 90', 'fouls per 90',
            'aerial duels per 90', 'successful defensive actions per 90',
        ]

        def_off_cols = [
            'shots per 90', 'crosses per 90', 'dribbles per 90', 'offensive duels per 90',
            'touches in box per 90', 'progressive runs per 90', 'accelerations per 90',
            'received passes per 90', 'received long passes per 90', 'fouls suffered per 90',
            'passes per 90', 'forward passes per 90', 'back passes per 90',
            'passes to final third per 90', 'progressive passes per 90',
            'lateral passes per 90', 'short / medium passes per 90',
            'long passes per 90', 'passes to penalty area per 90', 'through passes per 90',
            'deep completions per 90', 'deep completed crosses per 90',
        ]

        for col in def_def_cols:
            if col in self.df.columns:
                self.df[col] = self.df.apply(
                    lambda row: self.possession_adjust_def(row[col], row["poss_decimal"]),
                    axis=1
                )

        for col in def_off_cols:
            if col in self.df.columns:
                self.df[col] = self.df.apply(
                    lambda row: self.possession_adjust_off(row[col], row["poss_decimal"]),
                    axis=1
                )

        cols_to_round = def_def_cols + def_off_cols
        existing_cols = [col for col in cols_to_round if col in self.df.columns]
        self.df[existing_cols] = self.df[existing_cols].round(2)

        return self.df

# ===========================================
# RETRIEVE POSSESSION (FUZZY MATCH)
# ===========================================
class RetrievePossession:
    def __init__(self, df_possession: pd.DataFrame):
        """
        Tar inn en DataFrame med possession-data (fra FBref eller lignende)
        og forbereder den for matching.
        Forventer kolonnene: 'squad' (lag) og 'possession' (0–100).
        """
        self.df_poss = df_possession.copy()
        self.df_poss.columns = self.df_poss.columns.str.lower()

        if "possession" not in self.df_poss.columns:
            raise KeyError(
                "Forventer en kolonne som heter 'possession' i df_possession."
            )
        if "squad" not in self.df_poss.columns:
            raise KeyError(
                "Forventer en kolonne som heter 'squad' (lagenavn) i df_possession."
            )

        self.df_poss["poss_decimal"] = self.df_poss["possession"] / 100.0

    def match_teams(self, team_name, possibilities, threshold=80):
        """
        Bruker fuzzy matching for å finne nærmeste match for et lag.
        Hvis beste match har score under threshold, beholdes originalnavnet.
        """
        match = process.extractOne(team_name, possibilities)
        if match and match[1] >= threshold:
            return match[0]
        return team_name

    def merge_possession_data(self, processed_df: pd.DataFrame) -> pd.DataFrame:
        """
        Matcher lag i processed_df med df_poss, og legger til poss_decimal.
        Forventer at processed_df har en kolonne 'squad'.
        Returnerer en ny DataFrame.
        """
        if "squad" not in processed_df.columns:
            raise KeyError("Forventer en kolonne 'squad' i processed_df.")

        poss_teams = self.df_poss["squad"].unique()

        processed_df = processed_df.copy()
        processed_df["squad_matched"] = processed_df["squad"].apply(
            lambda x: self.match_teams(x, poss_teams)
        )

        merged_df = pd.merge(
            processed_df,
            self.df_poss[["squad", "poss_decimal"]],
            left_on="squad_matched",
            right_on="squad",
            how="left",
        )

        merged_df = merged_df.drop(columns=["squad_matched", "squad_y"])
        merged_df = merged_df.rename(columns={"squad_x": "squad"})

        return merged_df

# ===========================================
# SAFE_DIV + PLAYSTYLE-FEATURES
# ===========================================
def safe_div(a, b):
    """
    Trygg deling: returnerer 0 der nevner er 0 eller NaN.
    Fungerer for både pandas Series og numpy-arrays.
    """
    return np.where((b > 0) & (~pd.isna(b)), a / b, 0)

def add_playstyle_features(df: pd.DataFrame) -> pd.DataFrame:
    """
    Legger til alle ekstra "playstyle"-features som brukes i clustering og robustness.
    Brukes på tvers av 24/25, 23/24 og 25/26.
    """
    df = df.copy()

    # goals per 90 hvis den mangler
    if (
        'goals per 90' not in df.columns
        and 'goals' in df.columns
        and 'minutes played' in df.columns
    ):
        df['goals per 90'] = safe_div(df['goals'], df['minutes played']) * 90

    # xG per touch / shot i boks
    df['xg per box touch'] = safe_div(
        df.get('xg per 90', 0),
        df.get('touches in box per 90', 0)
    )

    df['xg per shot'] = safe_div(
        df.get('xg per 90', 0),
        df.get('shots per 90', 0)
    )

    # pasningsretning / struktur
    df['share_of_passes_forward%'] = safe_div(
        df.get('forward passes per 90', 0),
        df.get('passes per 90', 0)
    )

    df['received_long_pass_share'] = safe_div(
        df.get('received long passes per 90', 0),
        df.get('received passes per 90', 0)
    )

    df['forward_to_lateral_pass_ratio'] = safe_div(
        df.get('forward passes per 90', 0),
        df.get('lateral passes per 90', 0)
    )

    # duell-stil
    df['aerial_duels_share'] = safe_div(
        df.get('aerial duels per 90', 0),
        df.get('duels per 90', 0)
    )

    df['defensive_duels_share'] = safe_div(
        df.get('defensive duels per 90', 0),
        df.get('duels per 90', 0)
    )

    df['offensive_duels_share'] = safe_div(
        df.get('offensive duels per 90', 0),
        df.get('duels per 90', 0)
    )

    # interceptions-profil
    df['interception_share'] = safe_div(
        df.get('interceptions per 90', 0),
        df.get('successful defensive actions per 90', 0)
    )

    df['int_to_def_duels_ratio'] = safe_div(
        df.get('interceptions per 90', 0),
        df.get('defensive duels per 90', 0)
    )

    df['deep_completions_per_forward_pass'] = safe_div(
        df.get('deep completions per 90', 0),
        df.get('forward passes per 90', 0)
    )

    # bakover/lateral/framover-andeler
    df['share_of_passes_backwards%'] = safe_div(
        df.get('back passes per 90', 0),
        df.get('passes per 90', 0)
    ) * -1

    df['share_of_passes_lateral%'] = safe_div(
        df.get('lateral passes per 90', 0),
        df.get('passes per 90', 0)
    )

    # ballføring vs mottak
    df['carry_after_receive_ratio'] = safe_div(
        df.get('progressive runs per 90', 0),
        df.get('received passes per 90', 0)
    )
    df['carry_dribble_vs_pass_tilt'] = safe_div(
        df.get('progressive runs per 90', 0) + df.get('dribbles per 90', 0),
        df.get('passes per 90', 0)
    )

    # pasningstyper
    df['througball_ratio'] = safe_div(
        df.get('through passes per 90', 0),
        df.get('passes per 90', 0)
    )

    df['long_pass_ratio'] = safe_div(
        df.get('long passes per 90', 0),
        df.get('passes per 90', 0)
    )

    # dribleprofil
    df['dribble_ratio'] = safe_div(
        df.get('dribbles per 90', 0),
        df.get('received passes per 90', 0)
    )

    # boks og skuddtrykk
    df['box_presence_ratio'] = safe_div(
        df.get('touches in box per 90', 0),
        df.get('passes per 90', 0)
    )

    df['shot_ratio'] = safe_div(
        df.get('shots per 90', 0),
        df.get('passes per 90', 0)
    )

    df['key_passes_ratio'] = safe_div(
        df.get('key passes per 90', 0),
        df.get('passes per 90', 0)
    )

    # over/under-performance vs xG
    if 'goals per 90' in df.columns and 'xg per 90' in df.columns:
        df['o_u_performance_goals_xg'] = safe_div(
            df['goals per 90'],
            df['xg per 90']
        )

    # offensiv vs defensiv tilt
    df['midfield_tilt'] = safe_div(
        df.get('progressive passes per 90', 0) + df.get('progressive runs per 90', 0),
        df.get('defensive duels per 90', 0) + df.get('interceptions per 90', 0)
    )

    df['making_foul_ratio'] = safe_div(
        df.get('fouls per 90', 0),
        df.get('defensive duels per 90', 0)
    )

    df['being_fouled_ratio'] = safe_div(
        df.get('fouls suffered per 90', 0),
        df.get('offensive duels per 90', 0)
    )

    new_cols = [
        'xg per box touch',
        'xg per shot',
        'share_of_passes_forward%',
        'share_of_passes_backwards%',
        'share_of_passes_lateral%',
        'received_long_pass_share',
        'forward_to_lateral_pass_ratio',
        'aerial_duels_share',
        'defensive_duels_share',
        'offensive_duels_share',
        'interception_share',
        'int_to_def_duels_ratio',
        'deep_completions_per_forward_pass',
        'carry_after_receive_ratio',
        'througball_ratio',
        'long_pass_ratio',
        'dribble_ratio',
        'box_presence_ratio',
        'shot_ratio',
        'key_passes_ratio',
        'o_u_performance_goals_xg',
        'midfield_tilt',
        'making_foul_ratio',
        'being_fouled_ratio',
        'carry_dribble_vs_pass_tilt',
    ]

    existing_new_cols = [c for c in new_cols if c in df.columns]
    df[existing_new_cols] = df[existing_new_cols].round(2)

    return df

# ===========================================
# DEFINER ENDELIGE FEATURES (RUNDE 2)
# ===========================================
important_features = [
    'successful defensive actions per 90',
    'defensive duels per 90',
    'aerial duels per 90',
    'shots blocked per 90',
    'interceptions per 90',
    'making_foul_ratio',
    'interception_share',
    'defensive_duels_share',
    'offensive_duels_share',
    'received passes per 90',
    'received long passes per 90',
    'received_long_pass_share',
    'fouls suffered per 90',
    'average pass length, m',
    'average long pass length, m',
    'dribbles per 90',
    'offensive duels per 90',
    'progressive runs per 90',
    'accelerations per 90',
    'forward passes per 90',
    'back passes per 90',
    'lateral passes per 90',
    'long passes per 90',
    'passes to final third per 90',
    'passes to penalty area per 90',
    'through passes per 90',
    'deep completions per 90',
    'share_of_passes_forward%',
    'share_of_passes_backwards%',
    'share_of_passes_lateral%',
    'forward_to_lateral_pass_ratio',
    'long_pass_ratio',
    'througball_ratio',
    'key_passes_ratio',
    'touches in box per 90',
    'shot_ratio',
    'shots per 90',
    'midfield_tilt'
]

features_to_weight = [
    'touches in box per 90',
    'dribbles per 90',
    'progressive runs per 90',
    'share_of_passes_forward%',
    'share_of_passes_lateral%',
    'share_of_passes_backwards%',
    'passes to final third per 90',
    'passes to penalty area per 90',
    'long_pass_ratio',
    'through passes per 90',
    'defensive duels per 90',
    'midfield_tilt',
    'defensive_duels_share',
    'offensive_duels_share',
    'received passes per 90',
    'aerial duels per 90',
    'successful defensive actions per 90'
]

# ===========================================
# PREPROCESS 24/25 + POSSESSION + FEATURE-MATRIX
# ===========================================
leagues_top5 = [
    'Spain La Liga 2024-25',
    'Italy Serie A 2024-25',
    'France Ligue 1 2024-25',
    'Germany Bundesliga 2024-25',
    'England Premier League 2024-25'
]

positions = ['RCMF', 'LCMF']

pre_2425 = Preprocessing(
    df_24_25,
    positions=positions,
    min_minutes=900,
    leagues=leagues_top5 if not USE_EXAMPLE_DATA else None
)

filtered_df_2425 = pre_2425.filter_data()

context_cols_2425 = [
    'player', 'squad', 'club today', 'age', 'market value',
    'position', 'contract months left', 'minutes played'
]
context_data_2425 = filtered_df_2425[context_cols_2425].copy()

# Possession-justering via RetrievePossession (hvis mulig)
if not USE_EXAMPLE_DATA and df_possession is not None:
    try:
        poss_handler = RetrievePossession(df_possession)
        merged_df_2425 = poss_handler.merge_possession_data(filtered_df_2425)
        adjuster = NormalizePossession(merged_df_2425)
        df_2425_poss_adj = adjuster.apply_adjustments()
        print("✅ Possession-justering brukt for 24/25.")
    except Exception as e:
        print(f"⚠️ Possession-justering hoppes over pga feil: {e}")
        df_2425_poss_adj = filtered_df_2425.copy()
else:
    df_2425_poss_adj = filtered_df_2425.copy()

# Numeriske kolonner for 24/25
df_clean_2425 = df_2425_poss_adj.select_dtypes(include=['number']).copy()

# Legg på playstyle-features
df_features_2425 = add_playstyle_features(df_clean_2425)

# ===========================================
# RUNDE 1 – FEATURE SELECTION + EXPLORATORY CLUSTERING
# ===========================================
features_round1 = [
    'successful defensive actions per 90', 'defensive duels per 90',
    'aerial duels per 90', 'shots blocked per 90', 'interceptions per 90',
    'fouls per 90', 'fouls suffered per 90',
    'duels per 90', 'offensive duels per 90',
    'dribbles per 90', 'touches in box per 90',
    'progressive runs per 90', 'accelerations per 90',
    'received passes per 90', 'received long passes per 90',
    'passes per 90', 'forward passes per 90', 'back passes per 90',
    'lateral passes per 90', 'short / medium passes per 90',
    'long passes per 90',
    'passes to final third per 90', 'passes to penalty area per 90',
    'through passes per 90', 'deep completions per 90',
    'deep completed crosses per 90',
    'xg per 90', 'shots per 90', 'key passes per 90',
    'share_of_passes_forward%', 'share_of_passes_backwards%', 'share_of_passes_lateral%',
    'received_long_pass_share', 'forward_to_lateral_pass_ratio',
    'aerial_duels_share', 'defensive_duels_share', 'offensive_duels_share',
    'interception_share', 'int_to_def_duels_ratio',
    'deep_completions_per_forward_pass', 'carry_after_receive_ratio',
    'througball_ratio', 'long_pass_ratio', 'dribble_ratio',
    'box_presence_ratio', 'shot_ratio', 'key_passes_ratio',
    'midfield_tilt', 'making_foul_ratio', 'being_fouled_ratio',
    'carry_dribble_vs_pass_tilt',
]

features_round1 = [f for f in features_round1 if f in df_features_2425.columns]

# Korrelasjon
corr_matrix = df_features_2425[features_round1].corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop_corr = [column for column in upper.columns if any(upper[column] > 0.9)]
features_corr_filtered = [f for f in features_round1 if f not in to_drop_corr]

# Varians
X_round1 = df_features_2425[features_corr_filtered]
mm_scaler = MinMaxScaler()
X_round1_scaled = mm_scaler.fit_transform(X_round1)
X_round1_scaled_df = pd.DataFrame(X_round1_scaled, columns=features_corr_filtered)

variances = X_round1_scaled_df.var()
low_var = variances[variances < 0.02].index.tolist()
features_round1_final = [f for f in features_corr_filtered if f not in low_var]

print("Antall features etter korrelasjon+varians i runde 1:", len(features_round1_final))

# PCA + KMeans runde 1
X1 = df_features_2425[features_round1_final]
scaler_round1 = StandardScaler()
X1_scaled = scaler_round1.fit_transform(X1)

pca_round1 = PCA(n_components=0.90)
X_pca_round1 = pca_round1.fit_transform(X1_scaled)
print(f"Runde 1 – antall PCA-komponenter: {X_pca_round1.shape[1]}")

# Plot forklart varians runde 1
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(pca_round1.explained_variance_ratio_)+1),
         pca_round1.explained_variance_ratio_,
         marker='o', linestyle='--')
plt.title('Runde 1 – Explained variance per PCA-component')
plt.xlabel('Component number')
plt.ylabel('Explained variance')
plt.grid(True)
plt.gca().set_facecolor('#f0f0f0')
plt.tight_layout()
plt.savefig(VIZ_DIR / "runde1_PCA_explained_variance_per_component.png",
            dpi=300, bbox_inches='tight')
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(pca_round1.explained_variance_ratio_)+1),
         pca_round1.explained_variance_ratio_.cumsum(),
         marker='o')
plt.axhline(0.90, color='r', linestyle='--', label='90% variance')
plt.title('Runde 1 – Cumulative explained variance')
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
plt.legend()
plt.grid(True)
plt.gca().set_facecolor('#f0f0f0')
plt.tight_layout()
plt.savefig(VIZ_DIR / "runde1_PCA_cumulative_explained_variance.png",
            dpi=300, bbox_inches='tight')
plt.show()

# Elbow for runde 1
inertia = []
k_range = range(1, 11)
for k in k_range:
    km = KMeans(n_clusters=k, random_state=42, n_init='auto')
    km.fit(X_pca_round1)
    inertia.append(km.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker='o')
plt.title('Runde 1 – Elbow Method after PCA')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.xticks(k_range)
plt.grid(True)
plt.gca().set_facecolor('#f0f0f0')
plt.tight_layout()
plt.savefig(VIZ_DIR / "runde1_elbow_method_pca.png",
            dpi=300, bbox_inches='tight')
plt.show()

# Silhouette (runde 1 – optional)
scores = []
for k in range(2, 11):
    km = KMeans(n_clusters=k, random_state=42, n_init='auto')
    labels_k = km.fit_predict(X_pca_round1)
    score = silhouette_score(X_pca_round1, labels_k)
    scores.append(score)

plt.figure(figsize=(8, 5))
plt.plot(range(2, 11), scores, marker='o')
plt.title("Runde 1 – Silhouette Score for different k")
plt.xlabel("Number of clusters")
plt.ylabel("Silhouette Score")
plt.grid(True)
plt.gca().set_facecolor('#f0f0f0')
plt.tight_layout()
plt.savefig(VIZ_DIR / "runde1_silhouette_scores.png",
            dpi=300, bbox_inches='tight')
plt.show()

# KMeans runde 1 (velg k etter elbow/silhouette – her f.eks. 6)
k_round1 = 6
kmeans_round1 = KMeans(n_clusters=k_round1, random_state=42, n_init='auto')
clusters_round1 = kmeans_round1.fit_predict(X_pca_round1)

df_clusters_round1 = context_data_2425.copy()
df_clusters_round1['cluster_round1'] = clusters_round1

# RF feature importance runde 1
rf_round1 = RandomForestClassifier(random_state=42)
rf_round1.fit(X1_scaled, df_clusters_round1['cluster_round1'])

importances_round1 = pd.Series(rf_round1.feature_importances_, index=features_round1_final)
importances_round1_sorted = importances_round1.sort_values(ascending=False)

top20_r1 = importances_round1_sorted.head(20)
norm_r1 = (top20_r1 - top20_r1.min()) / (top20_r1.max() - top20_r1.min())
colors_r1 = sns.color_palette("crest", as_cmap=True)(norm_r1)

plt.figure(figsize=(10, 6))
ax = sns.barplot(x=top20_r1, y=top20_r1.index, palette=colors_r1)
ax.set_facecolor('#f0f0f0')
plt.grid(True, axis='x', linestyle='--', alpha=0.5)
plt.title('Runde 1 – Top 20 features (Random Forest)', fontsize=14)
plt.xlabel('Feature importance')
plt.ylabel('Feature')
plt.tight_layout()
plt.savefig(VIZ_DIR / "runde1_RF_top20_features.png",
            dpi=300, bbox_inches='tight')
plt.show()

print("Runde 1 – Top 20 features:\n", top20_r1)

# ===========================================
# RUNDE 2 – ENDELIG MODELL (BASELINE 24/25)
# ===========================================
def fit_baseline_2425(df_features_2425: pd.DataFrame,
                      context_data_2425: pd.DataFrame,
                      important_features,
                      features_to_weight):
    """
    Endelig modell (runde 2):
    - bruker 'important_features'
    - vekter 'features_to_weight'
    - returnerer alt vi trenger for å projisere andre sesonger.
    """
    df_num = df_features_2425.copy()

    important_2425 = [c for c in important_features if c in df_num.columns]
    X_2425 = df_num[important_2425].copy()

    # fyll NaN med mean (24/25)
    for col in important_2425:
        X_2425[col] = X_2425[col].fillna(X_2425[col].mean())

    scaler = StandardScaler().fit(X_2425)
    X_scaled = pd.DataFrame(
        scaler.transform(X_2425),
        columns=important_2425,
        index=X_2425.index
    )

    # vekting (f.eks. *2)
    for feat in features_to_weight:
        if feat in X_scaled.columns:
            X_scaled[feat] = X_scaled[feat] * 2

    # PCA runde 2
    pca = PCA(n_components=0.90).fit(X_scaled)
    X_pca_weighted = pca.transform(X_scaled)
    print(f"Runde 2 – antall PCA-komponenter: {X_pca_weighted.shape[1]}")

    # plott PCA varians
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, len(pca.explained_variance_ratio_)+1),
             pca.explained_variance_ratio_,
             marker='o', linestyle='--')
    plt.title('Runde 2 – Explained variance per PCA-component (after weighting)')
    plt.xlabel('Component number')
    plt.ylabel('Explained variance')
    plt.grid(True)
    plt.gca().set_facecolor('#f0f0f0')
    plt.tight_layout()
    plt.savefig(VIZ_DIR / "runde2_PCA_explained_variance_per_component.png",
                dpi=300, bbox_inches='tight')
    plt.show()

    plt.figure(figsize=(10, 6))
    plt.plot(range(1, len(pca.explained_variance_ratio_)+1),
             pca.explained_variance_ratio_.cumsum(),
             marker='o')
    plt.axhline(0.90, color='r', linestyle='--', label='90% variance')
    plt.title('Runde 2 – Cumulative explained variance (after weighting)')
    plt.xlabel('Number of components')
    plt.ylabel('Cumulative explained variance')
    plt.legend()
    plt.grid(True)
    plt.gca().set_facecolor('#f0f0f0')
    plt.tight_layout()
    plt.savefig(VIZ_DIR / "runde2_PCA_cumulative_explained_variance.png",
                dpi=300, bbox_inches='tight')
    plt.show()

    # Elbow runde 2
    inertia_r2 = []
    k_range = range(1, 11)
    for k in k_range:
        km = KMeans(n_clusters=k, random_state=42, n_init='auto')
        km.fit(X_pca_weighted)
        inertia_r2.append(km.inertia_)

    plt.figure(figsize=(8, 5))
    plt.plot(k_range, inertia_r2, marker='o')
    plt.title('Runde 2 – Elbow Method after PCA with weightings')
    plt.xlabel('Number of clusters (k)')
    plt.ylabel('Inertia')
    plt.xticks(k_range)
    plt.grid(True)
    plt.gca().set_facecolor('#f0f0f0')
    plt.tight_layout()
    plt.savefig(VIZ_DIR / "runde2_elbow_method_pca_weighted.png",
                dpi=300, bbox_inches='tight')
    plt.show()

    # Velg endelig k (her 5)
    optimal_k = 5
    kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init='auto')
    cluster_labels = kmeans.fit_predict(X_pca_weighted)

    df_2425_clusters = context_data_2425.copy()
    df_2425_clusters['cluster_2425'] = cluster_labels

    # RF – endelig feature importance
    rf_final = RandomForestClassifier(random_state=42)
    rf_final.fit(X_scaled, cluster_labels)
    importances_final = pd.Series(rf_final.feature_importances_, index=important_2425).sort_values(ascending=False)

    # Plott top 20 final features
    top20_final = importances_final.head(20)
    norm_f = (top20_final - top20_final.min()) / (top20_final.max() - top20_final.min())
    colors_f = sns.color_palette("crest", as_cmap=True)(norm_f)

    plt.figure(figsize=(10, 6))
    ax = sns.barplot(x=top20_final, y=top20_final.index, palette=colors_f)
    ax.set_facecolor('#f0f0f0')
    plt.grid(True, axis='x', linestyle='--', alpha=0.5)
    plt.title('Runde 2 – Top 20 features explaining clusters', fontsize=14)
    plt.xlabel('Feature importance')
    plt.ylabel('Feature')
    plt.tight_layout()
    plt.savefig(VIZ_DIR / "runde2_RF_top20_features.png",
                dpi=300, bbox_inches='tight')
    plt.show()

    print("Runde 2 – Top 20 features:\n", top20_final)

    return {
        "important_2425": important_2425,
        "X_2425": X_2425,
        "scaler": scaler,
        "pca": pca,
        "kmeans": kmeans,
        "df_2425_clusters": df_2425_clusters,
        "features_to_weight": features_to_weight,
        "X_pca_weighted": X_pca_weighted,
        "rf_final_importances": importances_final,
    }

baseline_2425 = fit_baseline_2425(
    df_features_2425=df_features_2425,
    context_data_2425=context_data_2425,
    important_features=important_features,
    features_to_weight=features_to_weight
)

# ===========================================
# VISUALISERING AV CLUSTERS (24/25) I PCA-ROM
# ===========================================
X_pca_weighted_2425 = baseline_2425["X_pca_weighted"]
df_2425_clusters = baseline_2425["df_2425_clusters"]

pca_df_2425 = pd.DataFrame(
    X_pca_weighted_2425[:, :2],
    columns=['PCA1', 'PCA2'],
    index=df_2425_clusters.index
)
pca_df_2425['cluster'] = df_2425_clusters['cluster_2425']
pca_df_2425['player'] = df_2425_clusters['player']

closest_players = {}
for cluster_label in pca_df_2425['cluster'].unique():
    cluster_points = pca_df_2425[pca_df_2425['cluster'] == cluster_label]
    center = cluster_points[['PCA1', 'PCA2']].mean().values
    distances = np.linalg.norm(cluster_points[['PCA1', 'PCA2']].values - center, axis=1)
    closest_idx = distances.argmin()
    closest_player = cluster_points.iloc[closest_idx]['player']
    closest_players[cluster_label] = closest_player

print("Closest-to-centre players per cluster (24/25):")
for c, p in closest_players.items():
    print(f"Cluster {c}: {p}")

custom_palette = {
    0: '#0099FF',
    1: '#E60000',
    2: '#ff99cc',
    3: '#33CCCC',
    4: '#00CC66',
}

label_players = ['A. Bianco', 'R. Khedira', 'N. El Aynaoui', 'O. Targhalline', 'C. Boukhalfa']

plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=pca_df_2425['PCA1'],
    y=pca_df_2425['PCA2'],
    hue=pca_df_2425['cluster'],
    palette=custom_palette
)
plt.title('Final clusters visualized by PCA (24/25)')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.grid(True)
ax = plt.gca()
ax.set_facecolor('#f0f0f0')

for player_name in label_players:
    player_row = pca_df_2425[pca_df_2425['player'] == player_name]
    if not player_row.empty:
        x = player_row['PCA1'].iloc[0]
        y = player_row['PCA2'].iloc[0]
        plt.text(x + 0.1, y + 0.1, player_name, fontsize=8, weight='bold', color='black',
                 bbox=dict(facecolor='white', edgecolor='none', alpha=0.7))

plt.tight_layout()
plt.savefig(VIZ_DIR / "runde2_clusters_PCA1_PCA2_labeled.png",
            dpi=400, bbox_inches='tight')
plt.show()

# ===========================================
# ROBUSTNESS FUNKSJON – PROJECT_SEASON_AND_REPORT
# ===========================================
def project_season_and_report(filtered_df_other: pd.DataFrame,
                              context_data_other: pd.DataFrame,
                              baseline: dict,
                              season_label: str,
                              colname_cluster_other: str,
                              colname_axis: str):
    """
    Projiserer en annen sesong (23/24 eller 25/26) inn i 24/25-modellen,
    printer stabilitet og plotter transition-matrise. Bruker samme
    PCA+KMeans+weightings som endelig modell (runde 2).
    """
    df_2425_clusters = baseline["df_2425_clusters"]
    important_2425   = baseline["important_2425"]
    X_2425           = baseline["X_2425"]
    scaler           = baseline["scaler"]
    pca              = baseline["pca"]
    kmeans           = baseline["kmeans"]
    weighted_feats   = baseline["features_to_weight"]

    df_other_num = filtered_df_other.select_dtypes(include=['number']).copy()
    df_other_num = add_playstyle_features(df_other_num)

    for col in important_2425:
        if col not in df_other_num.columns:
            df_other_num[col] = X_2425[col].mean()

    X_other = df_other_num[important_2425].copy()

    for col in important_2425:
        X_other[col] = X_other[col].fillna(X_2425[col].mean())

    X_other_scaled = pd.DataFrame(
        scaler.transform(X_other),
        columns=important_2425,
        index=X_other.index
    )

    for feat in weighted_feats:
        if feat in X_other_scaled.columns:
            X_other_scaled[feat] = X_other_scaled[feat] * 2

    X_other_pca_weighted = pca.transform(X_other_scaled)

    df_other_clusters = context_data_other.copy()
    df_other_clusters[colname_cluster_other] = kmeans.predict(X_other_pca_weighted)

    merged_roles = df_2425_clusters.merge(
        df_other_clusters[['player', colname_cluster_other]],
        on='player',
        how='inner'
    )

    role_stability = (merged_roles['cluster_2425'] == merged_roles[colname_cluster_other]).mean()
    print(f"Cross-season role stability (24/25 baseline vs {season_label} projected): {role_stability:.2%}")

    stability_per_role = (
        merged_roles
        .assign(same = merged_roles['cluster_2425'] == merged_roles[colname_cluster_other])
        .groupby('cluster_2425')['same']
        .mean()
        .rename('stability')
    )
    print(stability_per_role)

    pivot = pd.crosstab(
        merged_roles['cluster_2425'],
        merged_roles[colname_cluster_other],
        rownames=['2024/25 role'],
        colnames=[colname_axis]
    )

    plt.figure(figsize=(8, 6))
    sns.heatmap(pivot, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Role transition matrix (2024/25 baseline → {season_label} projected)')
    plt.tight_layout()
    fname = f"transition_{season_label.replace('/', '_')}.png"
    plt.savefig(VIZ_DIR / fname, dpi=300, bbox_inches='tight')
    plt.show()

# ===========================================
# ROBUSTNESS – 25/26
# ===========================================
context_cols_robust = [
    'player', 'squad', 'club today', 'age', 'market value',
    'position', 'contract months left', 'minutes played'
]

if df_25_26 is None:
    print("⚠️ Robustness test (25/26) hoppes over: df_25_26 er ikke tilgjengelig.")
else:
    pre_2526 = Preprocessing(
        df_25_26,
        positions=['RCMF', 'LCMF'],
        min_minutes=400,
        leagues=None
    )

    filtered_df_2526 = pre_2526.filter_data()
    context_data_2526 = filtered_df_2526[context_cols_robust].copy()

    project_season_and_report(
        filtered_df_other=filtered_df_2526,
        context_data_other=context_data_2526,
        baseline=baseline_2425,
        season_label="2025/26",
        colname_cluster_other="cluster_2526",
        colname_axis="2025/26 projected"
    )

# ===========================================
# ROBUSTNESS – 23/24
# ===========================================
if df_23_24 is None:
    print("⚠️ Robustness test (23/24) hoppes over: df_23_24 er ikke tilgjengelig.")
else:
    pre_2324 = Preprocessing(
        df_23_24,
        positions=['RCMF', 'LCMF'],
        min_minutes=900,
        leagues=None
    )

    filtered_df_2324 = pre_2324.filter_data()
    context_data_2324 = filtered_df_2324[context_cols_robust].copy()

    project_season_and_report(
        filtered_df_other=filtered_df_2324,
        context_data_other=context_data_2324,
        baseline=baseline_2425,
        season_label="2023/24",
        colname_cluster_other="cluster_2324",
        colname_axis="2023/24 projected"
    )
```
